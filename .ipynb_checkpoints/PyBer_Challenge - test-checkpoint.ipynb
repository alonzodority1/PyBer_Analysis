# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
type_ride_count_df=pyber_data_df.groupby(["type"]).count()["ride_id"]
type_ride_count_df
```

```python
# 2. Get the total drivers for each city type
type_driver_count_df=city_data_df.groupby(["type"]).sum()["driver_count"]
type_driver_count_df
```

```python
#  3. Get the total amount of fares for each city type
type_fare_count_df=pyber_data_df.groupby(["type"]).sum()["fare"]
type_fare_count_df
```

```python
#  4. Get the average fare per ride for each city type. 
type_FPR_average = type_fare_count_df / type_ride_count_df
type_FPR_average
```

```python
# 5. Get the average fare per driver for each city type. 
type_FPD_average = type_fare_count_df / type_driver_count_df
type_FPD_average
```

```python
#  6. Create a PyBer summary DataFrame. 

pyber_summary_df = pd.DataFrame(
          {"Total Rides": type_ride_count_df, 
          "Total Drivers": type_driver_count_df, 
          "Total Fares": type_fare_count_df,
          "Average Fair per Ride": type_FPR_average, 
          "Average Fair per Driver": type_FPD_average})




# Display the data frame
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fair per Ride"] = pyber_summary_df["Average Fair per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fair per Driver"] = pyber_summary_df["Average Fair per Driver"].map("${:,.2f}".format)

pyber_summary_df

```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
pyber_by_type_date=pyber_data_df.groupby(["type","date"]).sum()["fare"]
pyber_by_type_date
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
pyber_by_type_date=pyber_by_type_date.reset_index()
pyber_by_type_date
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pyber_by_type_date=pyber_by_type_date.pivot(index="date", columns="type", values="fare")
pyber_by_type_date
```

```python
pyber_by_type_date.describe()
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
trimestre_one_df=pyber_by_type_date.loc["2019-01-01" : "2019-04-29"]
trimestre_one_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
trimestre_one_df.index = pd.to_datetime(trimestre_one_df.index)
trimestre_one_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()

trimestre_one_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
trimester_one_per_week_df=trimestre_one_df.resample("w").sum()
trimester_one_per_week_df
```

```python
# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

trimester_one_per_week_df.plot.line(figsize=(20,7), title="Total Fare by City Type")
plt.legend(loc="center", title="type")
plt.ylabel("Fare ($USD)")
plt.xlabel("")

plt.savefig("analysis/Pyber_fare_DF_Method.png")

plt.show()
```

```python
# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

fig,ax = plt.subplots(figsize=(20,7))
ax.plot(trimester_one_per_week_df["Rural"], label = "Rural")
ax.plot(trimester_one_per_week_df["Suburban"], label = "Suburban")
ax.plot(trimester_one_per_week_df["Urban"], label = "Urban")

x_labels = trimester_one_per_week_df.index

locations =[x_labels[0], x_labels[4], x_labels[8], x_labels[13]]

labels=["Jan\n2019", "Feb", "Mar", "Apr"]

ax.set_xlim,(x_labels[0],x_labels[-1])

plt.xticks(locations,labels)

ax.set_title("Total Fare by City Type")
ax.legend(title= 'type', loc="center")
ax.set_ylabel("Fare($USD)")

plt.savefig("analysis/Pyber_fare_summary_Object_Method.png")

plt.show()


```
